```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
```

# Research on relese month and revenue

```python
#load the data and the column name
df = pd.read_csv('movie.metadata.tsv',sep='\t',header=None)
with open('namecolumn.txt', 'r') as file:
    column_name = file.read()
df.columns = column_name.split('\n')
#drop the nan for column of 'release date' and 'revenue'
df.dropna(subset=['Movie release date', 'Movie box office revenue'],inplace=True)
condition = df['Movie release date'].str.match(r'^\d{4}$')
```

```python
df
```

## Extract the month for the data and transfer them into dummy variables


```python
aux = df[~condition].copy(deep=True)
aux['month'] = aux['Movie release date'].str.extract(r'-(\d{2})-')
Y = aux['Movie box office revenue'].values
X = pd.get_dummies(aux['month']).values
X = X[:,:-1]
X.shape, Y.shape
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()
model.summary()
```

## Finding

Let Y refer to the random variable from sample of column 'Movie box office revenue'
Let X refer to the dummy variables of each month. X have size (N,1+11). N refer to the samples. '1' refer to the constant column with all elements are one. '11' refer the the dummy column from Jan to Nov (e.g. for one samples, if it is August, the element in column Aug will be 1, and 0 for other columns). 
Example:


If we run the regression of Y on X, it means we use December as base mark. We can see the coefficients in May, June, July and November are significant and positive. That means, on average, the 'Movie box office revenue' will be have higher value if we release movie on May, June, July and November than the other months.
